# Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
face = cv2.imread("mark.jpg")
face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
plt.imshow(face,cmap='gray')
```

# Face Detection 

```python
# Using Pre trained model using Vialo Jones Algorithm on Haar Cascade classifier
face_detec = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detect_fun( image ):
    face_copy = image.copy()
    face_rect = face_detec.detectMultiScale(face_copy)
    
    for (x,y,w,h) in face_rect: 
        cv2.rectangle(face_copy, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_copy
```

```python
plt.imshow(face_detect_fun(face),cmap='gray')
```

```python
# Detecting face in group 
group_pic = cv2.cvtColor(cv2.imread("group.jpg"), cv2.COLOR_BGR2GRAY)
plt.imshow(face_detect_fun(group_pic), cmap="gray")
```

```python
# Better Detection 
def better_detect_face(image):
    face_copy = image.copy()
  
    face_rect = face_detec.detectMultiScale(face_copy,scaleFactor=1.2, minNeighbors=5) 
    
    for (x,y,w,h) in face_rect: 
        cv2.rectangle(face_copy, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_copy
```

```python
group_pic = cv2.cvtColor(cv2.imread("group.jpg"), cv2.COLOR_BGR2GRAY)
plt.imshow(better_detect_face(group_pic), cmap="gray")
```

# Eye Detection

```python
eye_detect = cv2.CascadeClassifier('haarcascades/haarcascade_eye.xml')
```

```python
def eye_detect_fun( image ):
    eye_copy = image.copy()
    eye_rect = eye_detect.detectMultiScale(eye_copy, scaleFactor=1.2, minNeighbors=7)
    
    for (x,y,w,h) in eye_rect: 
        cv2.rectangle(eye_copy, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return eye_copy
```

```python
plt.imshow(eye_detect_fun(face),cmap='gray')
```

# Face and Eye Detection

```python
detected_face = eye_detect_fun(better_detect_face(face))
```

```python
plt.imshow(detected_face, cmap="gray")
```
